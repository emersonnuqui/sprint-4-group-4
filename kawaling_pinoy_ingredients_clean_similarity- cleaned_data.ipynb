```python
import itertools
import pandas as pd
#sample_clean_panlasang_pinoy_updated.csv
#reusing R's codes
df1 = pd.read_csv('sample_clean_kawaling_pinoy.csv')
data = df1[["food","ingredients_clean"]].copy().reset_index().rename(columns = {"ingredients_clean": "text"})

#placeholder for unique combinations
list_comb = []

#utilizing itertools, which would create tuples. Tuples refer to every unique combination
for pair in itertools.combinations(data.text.tolist(), r=2):
    list_comb.append(pair)

#transforming the list of tuples to a dataframe, which should contain all unique pair combinations
df = pd.DataFrame(list_comb, columns=['text1', 'text2'])

#joining Filipino dishes details given text
#join first details on text1
df = df.merge(data[['food', 'text']].rename(columns={'food':'food1', 'text':'text1'}), left_on='text1', right_on='text1')
#join first details on text2
df = df.merge(data[['food', 'text']].rename(columns={'food':'food2', 'text':'text2'}), left_on='text2', right_on='text2')
```

```python
df2 = df.copy()
df2
```

```python
def get_similarities(text1, text2):
    doc1 = nlp(text1)
    doc2 = nlp(text2)
    print(doc1.text, "|" , doc2.text, doc1.similarity(doc2))

    return doc1.similarity(doc2)
```

```python
df2['t1_t2_similarity'] = df2.apply(lambda x: get_similarities(x['text1'], x['text2']), axis = 1)
df2
```

```python
df2_sorted = df2.sort_values("t1_t2_similarity", ascending = False)
df2_sorted
```

```python
df2_sorted.to_csv("kaw_pinoy_similarity_scores_unsorted_cleaned_data.csv", index = False )
```
