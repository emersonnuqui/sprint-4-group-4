```python
import numpy as np
import pandas as pd
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("sample_clean_panlasang_pinoy.csv", index_col=0)
```

```python
df.shape
```

```python
df.head()
```

```python
#all ingredients
df["ingredients_clean"] = df["ingredients_clean"].astype(str)
text = " ".join(ingredients for ingredients in df.ingredients_clean)

# Create and generate a word cloud image:
wordcloud = WordCloud().generate(text)

# lower max_font_size, change the maximum number of word and lighten the background:
wordcloud = WordCloud(max_font_size=50, max_words=100, background_color="white").generate(text)
plt.figure(figsize=[10,10])
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.title("Most common ingredients in Filipino dishes", weight="bold")
plt.show()
```
